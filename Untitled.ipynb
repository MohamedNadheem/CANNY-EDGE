```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('images.webp')
```

```python
plt.subplot(1,1,1)
plt.title('Original')
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
blurred_img=cv2.blur(image,ksize=(3,3))
plt.imshow(blurred_img[:,:,::-1])
```

```python
medium=np.median(image)
```

```python
low=int(max(0,0.7*medium))
high=int(max(255,1/3*medium))
```

```python
canny_edges_blur=cv2.Canny(blurred_img,threshold1=low,threshold2=high)
plt.title('Canny Image')
plt.imshow(canny_edges_blur)
```
